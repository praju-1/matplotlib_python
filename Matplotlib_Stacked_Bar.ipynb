# Introduction

- It is a graph that is used to compare parts of a whole. 
- In a stacked bar chart each bar represents the whole, and the segments or parts in the bar represent categories of that whole. 
- Different colors are used to represent these categories.

# 1: (Simple stacked bar plot)

```python
# importing package
import matplotlib.pyplot as plt
 
# create data
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]
 
# plot bars in stack manner
plt.bar(x, y1, color='c')
plt.bar(x, y2, bottom=y1, color='r')
plt.show()
```

# 2: (Stacked bar chart with more than 2 data)

```python
# importing package
import matplotlib.pyplot as plt
import numpy as np
 
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
 
# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

# 3: (Stacked Bar chart using dataframe plot)

```python
# importing package
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.DataFrame([['IBM',105, 95, 66, 40], ['Infosys', 32, 30, 110, 26], ['Amazon', 25, 40, 20, 18],
                   ['Amdocs', 37, 18, 110, 19]],
                  columns=['Salary', 'January', 'February', 'March', 'April'])
# view data
print(df)
 
# plot data in stack manner of bar type
df.plot(x='Salary', kind='bar', stacked=True,
        title='Stacked Bar Graph by dataframe')
plt.show()
```

# Add Percentage on subgroups of each group.

```python
# importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
  
df = pd.DataFrame([['IBM',105, 95, 66, 40], ['Infosys', 32, 30, 110, 26], ['Amazon', 25, 40, 20, 18],
                   ['Amdocs', 37, 18, 110, 19]],
                  columns=['Salary', 'January', 'February', 'March', 'April'])

# view data
print(df)

# plot a Stacked Bar Chart using matplotlib
df.plot(
  x = 'Salary', 
  kind = 'barh', 
  stacked = True, 
  title = 'Percentage Stacked Bar Graph', 
  mark_right = True)

#calculte total of all salary amount
df_total = df["January"] + df["February"] + df["March"] + df["April"]

#Caluclate percentage of each column
df_rel = df[df.columns[1:]].div(df_total, 0)*100

#iterate through each percentage
for n in df_rel:
    for i, (cs, ab, pc) in enumerate(zip(df.iloc[:, 1:].cumsum(1)[n], 
                                         df[n], df_rel[n])):
        plt.text(cs - ab / 2, i, str(np.round(pc, 1)) + '%', 
                 va = 'center', ha = 'center')
```
