# * Matplotlib Plotting

# 1. Plotting x and y points

- The plot() function is used to draw points (markers) in a diagram.

- By default, the plot() function draws a line from point to point.

- The function takes parameters for specifying points in the diagram.

- Parameter 1 is an array containing the points on the x-axis.

- Parameter 2 is an array containing the points on the y-axis.

```python
# importing matplotlib module 
import matplotlib.pyplot as plt
  
# x-axis values
x = [5, 2, 9, 4, 7]
  
# Y-axis values
y = [7, 5, 8, 4, 2]
  
# Function to plot
plt.plot(x,y)
  
# function to show the plot
plt.show()
```

# 2. Plotting Without Line

- To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

```python
# importing matplotlib module 
import matplotlib.pyplot as plt
  
# x-axis values
x = [5, 2, 9, 4, 7]
  
# Y-axis values
y = [7, 5, 8, 4, 2]
  
# Function to plot
# we can provide any symbol here
plt.plot(x,y, 'o')
  
# function to show the plot
plt.show()
```

```python
# importing matplotlib module 
import matplotlib.pyplot as plt
   
# Y-axis values
y = [7, 5, 8, 4, 2]
  
# Function to plot
plt.plot(y)
  
# function to show the plot
plt.show()
```

# * Marker Size and Marker Color

- You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:

# 1. Create Labes for plot

- With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

```python
# importing matplotlib module 
import matplotlib.pyplot as plt

x = [1972, 1982, 1992, 2002, 2012, 2022]
y = [35, 37, 50, 43, 32, 42]

# Function to plot
plt.plot(x, y, marker = '*')

#creating labels
plt.xlabel("Year")
plt.ylabel("Temperature")

# function to show the plot
plt.show()
```

# 2. Create a Title for a Plot

```python
# importing matplotlib module 
import matplotlib.pyplot as plt

x = [1972, 1982, 1992, 2002, 2012, 2022]
y = [35, 37, 50, 43, 32, 42]

# Function to plot
plt.plot(x, y, marker = '*')

#creating labels
plt.title("Weather report")
plt.xlabel("Year")
plt.ylabel("Temperature")

# function to show the plot
plt.show()
```

# Set Font Properties for Title and Labels

- You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

- Also we can specify separate font to each label

```python
# importing matplotlib module 
import matplotlib.pyplot as plt

x = [1972, 1982, 1992, 2002, 2012, 2022]
y = [35, 37, 50, 43, 32, 42]

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

# Function to plot
plt.plot(x, y, marker = '*')

#creating labels
plt.title("Weather report", fontdict = font1)
plt.xlabel("Year", fontdict = font2)
plt.ylabel("Temperature", fontdict = font2)

# function to show the plot
plt.show()
```

# Position the Title

- You can use the loc parameter in title() to position the title.

- Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
# importing matplotlib module 
import matplotlib.pyplot as plt

x = [1972, 1982, 1992, 2002, 2012, 2022]
y = [35, 37, 50, 43, 32, 42]

# Function to plot
plt.plot(x, y, marker = '*')

#creating labels
plt.title("Weather report", loc = 'left')
plt.xlabel("Year", loc = 'right')
plt.ylabel("Temperature", loc = 'top')

# function to show the plot
plt.show()
```

# Matplotlib Subplot

- It includes
1. Display Multiple Plots
2. The subplot() Function
3. Title

# 1. Display Multiple Plots

- With the subplot() function you can draw multiple plots in one figure:

```python
#importing module
import matplotlib.pyplot as plt

#plot 1:
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

#plotting first plot
plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = [0, 1, 2, 3]
y = [10, 20, 30, 40]

#plotting second plot
plt.subplot(1, 2, 2)
plt.plot(x,y)

# function to show the plot
plt.show()
```

# 2. The subplot() Function

- The subplot() function takes three arguments that describes the layout of the figure.

- The layout is organized in rows and columns, which are represented by the first and second argument.

- The third argument represents the index of the current plot.

- plt.subplot(1, 2, 1)
- the figure has 1 row, 2 columns, and this plot is the first plot.

```python
#importing module
import matplotlib.pyplot as plt

x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

#Plotting first subplot
plt.subplot(2, 3, 1)
plt.plot(x,y)

x = [0, 1, 2, 3]
y = [10, 20, 30, 40]

#Plotting second subplot
plt.subplot(2, 3, 2)
plt.plot(x,y)

x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

#Plotting third subplot
plt.subplot(2, 3, 3)
plt.plot(x,y)

x = [0, 1, 2, 3]
y = [10, 20, 30, 40]

#Plotting fourth subplot
plt.subplot(2, 3, 4)
plt.plot(x,y)

plt.show()
```

# 3. Title

- You can add a title to each plot with the title() function:

```python
#importing module
import matplotlib.pyplot as plt

#plot 1:
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

#plotting first plot
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("INCOME")

#plot 2:
x = [0, 1, 2, 3]
y = [10, 20, 30, 40]

#plotting second plot
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("PROFIT")

# function to show the plot
plt.show()
```

# 4. Super Title

- You can add a title to the entire figure with the suptitle() function:

```python
#importing module
import matplotlib.pyplot as plt

#plot 1:
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

#plotting first plot
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("INCOME")

#plot 2:
x = [0, 1, 2, 3]
y = [10, 20, 30, 40]

#plotting second plot
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("PROFIT")

plt.suptitle("Month Review")
# function to show the plot
plt.show()
```

# 5. ColorMap

- The Matplotlib module has a number of available colormaps.

- A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.


- You can specify the colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.

```python
import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#colormap is like a list of colors, where each color has a value that ranges from 0 to 100.
colors = [0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100]

#plotting scatter plot
plt.scatter(x, y, c=colors, cmap='viridis')

#function to show the plot
plt.show()
```

- You can include the colormap in the drawing by including the plt.colorbar() statement:

```python
import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#colormap is like a list of colors, where each color has a value that ranges from 0 to 100.
colors = [0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100]

#plotting scatter plot
plt.scatter(x, y, c=colors, cmap='viridis')

#plotting colorbar
plt.colorbar()

#function to show the plot
plt.show()
```

# Matplotlib Histograms

- A histogram is a graph showing frequency distributions.

- It is a graph showing the number of observations within each given interval.

- In Matplotlib, we use the hist() function to create histograms.

- The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

```python
#importing module
import numpy as np
import matplotlib.pyplot as plt

#getting normal distribution with numpy method random
x = np.random.normal(170, 10, 250)

#create histogram
plt.hist(x)

#function to show plot
plt.show() 
```

# bin size

```python
#importing module
import numpy as np
import matplotlib.pyplot as plt

#getting normal distribution with numpy method random
x = np.random.normal(170, 10, 250)

#create histogram
plt.hist(x, bins = 20)

#function to show plot
plt.show() 
```

-  we can also provide other parameter to it

```python
#importing module
import numpy as np
import matplotlib.pyplot as plt

#getting normal distribution with numpy method random
x = np.random.normal(170, 10, 250)

#create histogram
plt.hist(x, bins = 20, color = 'r', density = True)

#function to show plot
plt.show() 
```

- With Pyplot, you can use the pie() function to draw pie charts:

- By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:

```python
#importing module
import matplotlib.pyplot as plt

y = [35, 25, 25, 15]

#create pie chart
plt.pie(y)

#function to show pie chart
plt.show() 
```

# 2. Label

- Add labels to the pie chart with the label parameter.

- The label parameter must be an array with one label for each wedge

```python
#importing module
import matplotlib.pyplot as plt

y = [35, 25, 25, 15]
mylabels = ["A", "B", "C", "D"]

#create pie chart
plt.pie(y, labels = mylabels)

#function to show pie chart
plt.show()
```

# 3. Start angle

- As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

- The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
#importing module
import matplotlib.pyplot as plt

y = [35, 25, 25, 15]
mylabels = ["A", "B", "C", "D"]

#create pie chart
plt.pie(y, labels = mylabels, startangle = 90)

#function to show pie chart
plt.show()
```

# 4. Explode

- Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.
- Each value represents how far from the center each wedge is displayed:

```python
#importing module
import matplotlib.pyplot as plt

y = [35, 25, 25, 15]
mylabels = ["A", "B", "C", "D"]

#give value to which you want to explode
myexplode = [0.2, 0, 0, 0]

#create pie chart
plt.pie(y, labels = mylabels, explode = myexplode)

#function to show pie chart
plt.show()
```

# 5. Shadow

 - Add a shadow to the pie chart by setting the shadows parameter to True:

```python
#importing module
import matplotlib.pyplot as plt

y = [35, 25, 25, 15]
mylabels = ["A", "B", "C", "D"]

#give value to which you want to explode
myexplode = [0.2, 0, 0, 0]

#create pie chart
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)

#function to show pie chart
plt.show()
```

# 6. colors

- You can set the color of each wedge with the colors parameter.

- The colors parameter, if specified, must be an array with one value for each wedge:

```python
#importing module
import matplotlib.pyplot as plt

y = [35, 25, 25, 15]
mylabels = ["A", "B", "C", "D"]

#give color for each part
mycolors = ["black", "chocolate", "grey", "#4CAF50"]


#create pie chart
plt.pie(y, labels = mylabels, colors = mycolors)

#function to show pie chart
plt.show()
```

# 7. Legend

- To add a list of explanation for each wedge, use the legend() function:

```python
#importing module
import matplotlib.pyplot as plt

y = [35, 25, 25, 15]
mylabels = ["A", "B", "C", "D"]

#create pie chart
plt.pie(y, labels = mylabels)

plt.legend()

#function to show pie chart
plt.show()
```

- We also provide column wise legend
- Bydefault it is row wise

```python
#importing module
import matplotlib.pyplot as plt

y = [35, 25, 25, 15]
mylabels = ["A", "B", "C", "D"]

#create pie chart
plt.pie(y, labels = mylabels)

plt.legend(ncol = 4)

#function to show pie chart
plt.show()
```

- We can also specifie font size, color

```python
#importing module
import matplotlib.pyplot as plt

y = [35, 25, 25, 15]
mylabels = ["A", "B", "C", "D"]

#create pie chart
plt.pie(y, labels = mylabels)

plt.legend(ncol = 4, facecolor = "yellow", fontsize = 15)

#function to show pie chart
plt.show()
```

# 8. Legend With Header

- To add a header to the legend, add the title parameter to the legend function.

```python
#importing module
import matplotlib.pyplot as plt

y = [35, 25, 25, 15]
mylabels = ["A", "B", "C", "D"]

#create pie chart
plt.pie(y, labels = mylabels)

plt.legend(title = "Category")

#function to show pie chart
plt.show()
```
